```python
import numpy as np
import matplotlib.pyplot as plt

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 1. 定义计算函数
def calculate_pi(n):
    """用内接正n边形近似计算圆周率π"""
    return n * np.sin(np.pi / n)

# 2. 计算不同n值下的π近似值和误差
n_values = np.array([1, 2, 4, 8, 16, 32, 64, 128, 256])
pi_approx = calculate_pi(n_values)
pi_exact = np.pi
error = np.abs(pi_exact - pi_approx)
h = 1 / n_values  # 单元尺寸h

# 3. 绘制双对数收敛图
plt.figure(figsize=(8, 6))

# 绘制误差点和连线
plt.loglog(h, error, 'o-', linewidth=2, markersize=8, label='数值解误差 $e_n$')

# 绘制参考线：斜率为2（二阶收敛）和斜率为4（四阶收敛）
# 参考线公式：log(e) = k*log(h) + b
h_ref = np.array([1e-3, 1e0])
# 斜率为2的参考线
e_ref2 = 10 * h_ref**2
plt.loglog(h_ref, e_ref2, 'k--', linewidth=1.5, label='斜率=2 (二阶收敛)')
# 斜率为4的参考线
e_ref4 = 100 * h_ref**4
plt.loglog(h_ref, e_ref4, 'k--', linewidth=1.5, label='斜率=4 (四阶收敛)')



for x, y, text in annotations:
    plt.annotate(text, (x, y), xytext=(5, 5), textcoords='offset points', fontsize=10)

# 4. 图表美化
plt.xlabel('单元尺寸 $h=1/n$', fontsize=12)
plt.ylabel('误差 $e_n=|\\pi-\\pi_n|$', fontsize=12)
plt.title('有限元法收敛性分析：圆周率π计算', fontsize=14)
plt.grid(True, which='both', linestyle='--', alpha=0.7)
plt.legend(fontsize=10)
plt.xlim(1e-3, 1e0)
plt.ylim(1e-10, 1e3)

# 5. 输出数值结果表格
print("="*60)
print("n\tπ近似值\t\t\t误差e_n\t\t\th=1/n")
print("="*60)
for n, pi, e, h_val in zip(n_values, pi_approx, error, h):
    print(f"{n}\t{pi:.15f}\t{e:.15f}\t{h_val:.6f}")
print("="*60)
print(f"精确值π: {pi_exact:.15f}")

plt.tight_layout()
plt.show()
```
